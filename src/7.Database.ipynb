```python
# Se guarda informacion en base de datos sqlite

import pandas as pd
import sqlite3

# Leer el archivo CSV
df = pd.read_csv('c://users//Sergi//proyectos//proyecto_final//data//processed//vehicles_with_model_clean.csv')

# Conectar a la base de datos (si no existe, se creará)
conn = sqlite3.connect('vehicles.db')

# Cargar los datos del DataFrame a la base de datos, en una tabla llamada 'vehicles'
df.to_sql('vehicles', conn, if_exists='replace', index=False)

# Cerrar la conexión
conn.close()

print("Datos cargados exitosamente en la tabla 'vehicles'.")
```

```python
import sqlite3

# Conectar a la base de datos y ejecutar la consulta SELECT
with sqlite3.connect('vehicles.db') as conn:
    cur = conn.cursor()
    cur.execute('SELECT DISTINCT model FROM vehicles '
                'WHERE manufacturer = "toyota"')
    rows = cur.fetchall()

# Mostrar los resultados
for row in rows:
    print(row)
```
